```python
import math
import matplotlib.pyplot as plt

def binomial_coefficient(n, k):
    return math.factorial(n) // (math.factorial(k) * math.factorial(n - k))

def calculate_sum(N, p):
    sum_n_greater_k = []
    sum_n_less_k = []
    for k in range(N + 1):
        term = binomial_coefficient(N, k) * (p ** k) * ((1 - p) ** (N - k))
        if N > k:
            sum_n_greater_k.append(term)
        else:
            sum_n_less_k.append(term)
    return sum_n_greater_k, sum_n_less_k

# Example usage
N = 5
p = 0.6
sum_n_greater_k, sum_n_less_k = calculate_sum(N, p)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(range(len(sum_n_greater_k)), sum_n_greater_k, label='n > k', linestyle='-', linewidth=2)
plt.plot(range(len(sum_n_less_k)), sum_n_less_k, label='n < k', linestyle='--', linewidth=2)
plt.plot(range(len(sum_n_greater_k)), [1 - x for x in sum_n_greater_k], label='1 - (n > k)', linestyle='-.', linewidth=2)
plt.plot(range(len(sum_n_less_k)), [1 - x for x in sum_n_less_k], label='1 - (n < k)', linestyle=':', linewidth=2)
plt.xlabel('k')
plt.ylabel('Sum')
plt.title('Sum of (N choose K) * p^k * (1-p)^(N-k)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import math
import matplotlib.pyplot as plt

def binomial_coefficient(n, k):
    return math.factorial(n) // (math.factorial(k) * math.factorial(n - k))

def calculate_sum(N, p):
    sum_n_greater_k = []
    sum_n_less_k = []
    max_n_greater_k = []
    for k in range(N + 1):
        term = binomial_coefficient(N, k) * (p ** k) * ((1 - p) ** (N - k))
        if N > k:
            sum_n_greater_k.append(term)
            max_n_greater_k.append(max(sum_n_greater_k))
        else:
            sum_n_less_k.append(term)
    return sum_n_greater_k, sum_n_less_k, max_n_greater_k

# Varying values of N and p
N_values = [30]
p_values = [x for x in range(1,500,5)]


plt.figure(figsize=(12, 8))
max_ = []

# Plotting for each combination of N and p
for N in N_values:
    for p in p_values:
        p = p / 1000.0
        sum_n_greater_k, sum_n_less_k, max_n_greater_k = calculate_sum(N, p)
        plt.plot(range(len(sum_n_greater_k)), sum_n_greater_k, label=f'N={N}, p={p} (n > k)')
        # plt.plot(range(len(sum_n_less_k)), sum_n_less_k, label=f'N={N}, p={p} (n < k)', linestyle='--')
        # plt.plot(range(len(sum_n_greater_k)), [1 - x for x in sum_n_greater_k], label=f'N={N}, p={p} (1 - n > k)', linestyle='-.')
        # plt.plot(range(len(sum_n_less_k)), [1 - x for x in sum_n_less_k], label=f'N={N}, p={p} (1 - n < k)', linestyle=':')
        plt.plot(range(len(max_n_greater_k)), max_n_greater_k, label=f'N={N}, p={p} (n > k)',linestyle=':')
        max_.append(max(max_n_greater_k))

plt.xlabel('k')
plt.ylabel('Sum')
plt.title('Sum of (N choose K) * p^k * (1-p)^(N-k)')
# plt.legend()
plt.grid(True)
plt.show()
```
